```python
import os
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
plt.interactive(False)
```

```python
histogram -> data grouped into bins
plt.hist (values, bins = n, normed=True, histtype="step")
```

```python
lets you discover, and show, the underlying frequency distribution (shape) of a set of continuous data.
good to check: shape of distribution, span, outliers
```

```python
histogram arguments
* bins (integer):  number of  intervals or bins
# a good number of bins is the square root of the number of data points
* range (tuple):  extrema of bins  (minimum,  maximum)
* normed  (boolean):  whether to  normalize to one
* cumulative  (boolean):  compute  Cumulative  Distribution Function (CDF)
```

```python
#DataFrame contendo 5 Séries com Distribuições Normais distintas
df = pd.DataFrame(columns=["S1", "S2", "S3", "S4", "S5"],
    data=(np.random.randn(100, 5) * np.array([10, 15, 50, 100, 200]) + np.array([0, 5, 30, 30, 50])))
```

```python
#sem Normalização: eixo Y é a contagem de elementos dentro de cada `bucket`.
plt.figure(figsize=(12,8))
plt.hist(df.S1, bins=10)
plt.show()
```

```python
#com Normalização: eixo Y é o percentual de elementos dentro de cada `bucket`.
plt.figure(figsize=(12,8))
plt.hist(df.S1, bins=10, normed=True)

plt.show()
```

```python
df.S1.describe()
```

### várias séries

```python
# dificil visualizar histograma de várias séries no matplotlib
#deve-se configurar o alpha (transparência) abaixo de 1.0 para histograma não sobrepor outro

plt.figure(figsize=(15,10))

plt.hist(df.S1, bins=10, normed=True, color="blue", alpha=0.5, label="S1")
plt.hist(df.S2, bins=10, normed=True, color="red", alpha=0.5, label="S2")
plt.legend()
plt.show()
```

```python
#usando plot do pandas histogramas saem separados
df.hist(bins=10, figsize=(12,8))

plt.show()
df.describe()
```

```python
#usando seaborn
plt.figure( figsize=(12,8))
sns.distplot(df.S1)
plt.show()
```

```python
plt.figure( figsize=(12,8))

f, axes = plt.subplots(3, 2, figsize=(15, 8), sharex=True)

sns.distplot(df.S1, kde=False, color="blue", ax=axes[0, 0])
sns.distplot(df.S2, kde=True, color="red", ax=axes[0, 1])
sns.distplot(df.S3, kde=True, color="orange", ax=axes[1, 0])
sns.distplot(df.S4, kde=True, rug=True, color="gray", ax=axes[1, 1])
sns.distplot(df.S5, hist=False, kde_kws={"shade": True}, color="purple", ax=axes[2, 1])

plt.show()
```
