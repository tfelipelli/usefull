```python
import os
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
#import seaborn as sns
```

### basic settings

```python
#essa linha permite plotar em um notebook
%matplotlib inline
#configurando o matplotlib para o modo manual
plt.interactive(False)

#sns.set()
```

### something to plot

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sin(x)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x * 0.333334
y4 = np.tan(x)
y5 = np.cosh(x)
```

### labels

```python
plt.xlabel("eixo X")
plt.ylabel("eixo Y")
plt.title("sen√≥ide")

plt.plot(x,y)
plt.show()
```

### font type, size, color...

```python
font = {'family': 'serif','color':  'darkred','weight': 'bold','size': 20}

plt.text(x=0, y=0, s="Texto a ser mostrado", fontdict=font)

plt.xlabel("eixo X", fontdict=font)
plt.ylabel("eixo Y", fontdict=font)
plt.title("sen√≥ide", fontdict=font)

plt.plot(x,y)
plt.show()
```

### size

```python
#12 por 8 polegadas
plt.figure(figsize=(12, 8))

plt.xlabel("eixo X")
plt.ylabel("eixo Y")
plt.title("sen√≥ide")

plt.plot(x,y)
plt.show()
```

```python
#redimensionando a janela de visualiza√ß√£o
plt.figure(figsize=(12, 8))
plt.xlim(-np.pi, np.pi)
plt.ylim(-1.2, +1.2)

plt.xlabel("eixo X")
plt.ylabel("eixo Y")
plt.title("sen√≥ide")

plt.plot(x,y)
plt.show()
```

### scale

```python
plt.xscale('log')

plt.plot(x,y)
plt.show()
```

### looks

```python
#keeps data off plot edges 2%
plt.margins(0.02) 

plt.figure(figsize=(12, 8))
plt.xlim(-np.pi, np.pi)
plt.ylim(-1.2, +1.2)

plt.xlabel("eixo X")
plt.ylabel("eixo Y")
plt.title("sen√≥ide")

plt.grid(True)

plt.plot(x,y)
plt.show()
```

### markers & colors

```python
plt.figure(figsize=(12, 8))
plt.xlim(-np.pi, np.pi)
plt.ylim(-1.2, +1.2)

plt.xlabel("eixo X")
plt.ylabel("eixo Y")
plt.title("sen√≥ide")

plt.grid(True)

plt.plot(x, y, marker='.', color="r", linestyle='none')
plt.show()
```

```python
#color: b (blue), r (red), g (green), c (cyan)
#marker: o (circle), * (star), s (square), + (plus)
#line: : (dotted), - (dashed)
```

### v√°rios plots juntos e legenda

```python
#construindo o plot
plt.figure(figsize=(12, 8))

plt.xlabel("eixo X")
plt.ylabel("eixo Y")
plt.title("sen√≥ide")

plt.xlim(-np.pi, np.pi)
plt.ylim(-1.2, +1.2)

plt.grid(True)

#plt.plot(x, y1, "r-", x, y2, "g--", x, y3, "b:")

plt.plot(x, y1, color="red", linewidth=1, linestyle="-")
plt.plot(x, y2, color="green", linewidth=2, linestyle="--")
plt.plot(x, y3, color="blue", linewidth=3, linestyle=":")
plt.plot(x, y4, color="darkorange", linewidth=4, linestyle="--")
plt.plot(x, y5, color="k", linewidth=5, linestyle="-")

local = "best"
#local = "upper right"
#local = "upper left"
#local = "lower left"
#local = "lower right"
#local = "right"
#local = "center left"
#local = "center right"
#local = "lower center"
#local = "upper center"
#local = "center"

plt.legend(["seno", "cosseno", "reta", "tangente", "cosseno hiperb√≥lico"])
```
